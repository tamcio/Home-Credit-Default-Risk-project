```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

from src.models.tuning import tune_model, tune_and_evaluate, build_best_pipeline
```

```python
from src.data_loader import merge_data
df = merge_data('../home-credit-default-risk')

df['TARGET'].value_counts(normalize=True)
```

```python
def prepare_data(df, drop=False, threshold=0.5):
    df_temp = df.copy()
    
    if drop:
        missing = df_temp.isna().mean()
        drop_cols = missing[missing > threshold].index
        df_temp = df_temp.drop(columns=drop_cols)
    
    y = df_temp['TARGET']
    X = df_temp.drop('TARGET', axis=1)
    
    X = pd.get_dummies(X, dummy_na=True)
    
    return X, y

X, y = prepare_data(df)
```

## Logistic Regression


```python
results_lr = tune_and_evaluate(
    X, y,
    model_type='logistic',
    n_trials=10,
    n_folds=3
)
```

## Random Forest

```python
results_rf = tune_and_evaluate(
    X, y,
    model_type='random_forest',
    n_trials=20,
    n_folds=3
)
```

## Gradient Boosting

```python
results_gb = tune_and_evaluate(
    X, y,
    model_type='gradient_boosting',
    n_trials=2,
    n_folds=3,
    timeout=1800
)
```

## XGBoost

```python
import xgboost

results_xgb = tune_and_evaluate(
    X, y,
    model_type='xgboost',
    n_trials=30,
    n_folds=3,
    use_gpu=True
)
```

## LightGBM

```python
import lightgbm

results_lgbm = tune_and_evaluate(
    X, y,
    model_type='lightgbm',
    n_trials=30,
    n_folds=3,
    use_gpu=True
)
```

## Compared Results

```python
results = {
    'Logistic': locals().get('results_lr'),
    'Random Forest': locals().get('results_rf'),
    'Gradient Boosting': locals().get('results_gb'),
    'XGBoost': locals().get('results_xgb'),
    'LightGBM': locals().get('results_lgbm'),
}

rows = []
for name, r in results.items():
    rows.append({
        'Model': name,
        'CV Score': r['cv_score'],
        'Train Score': r['train_score'],
        'Test Score': r['test_score']
    })

comp = pd.DataFrame(rows).sort_values('Test Score', ascending=False)
display(comp)
```
